```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0])
y_train = np.array([300,500])

m=len(x_train)
print(f"Number of training example is: {m}")
```

```python
i=0

x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}),y^({i}) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train,y_train)
plt.title("Housing Prices")
plt.ylabel("Price")
plt.xlabel("Size of House (sq.ft)")
plt.show()
```

```python
w = 100
b = 100

def calculate_model_output(w,b,x):
    m = x.shape
    f_wb = np.zeros(m)
    for i in range(len(x)):
        f_wb[i] = w*x[i] + b

    return f_wb
```

```python
tmp_f_wb = calculate_model_output(w,b,x_train)

plt.plot(x_train,tmp_f_wb,c='b',label='Our Prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='Actual Data')

plt.title("Housing Prices")
plt.ylabel("Price")
plt.xlabel("Size of House (sq.ft)")
plt.legend()
plt.show()
```
